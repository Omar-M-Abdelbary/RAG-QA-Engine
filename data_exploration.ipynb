```python
import pandas as pd

df = pd.read_csv(r'Natural-Questions-Base\Natural-Questions-Base.csv')
df.head()
```

```python
df['long_answers'].iloc[0]
df['short_answers'].iloc[0]
```

```python
df_clean = df[
    df['long_answers'].notnull() | 
    df['short_answers'].notnull()
]
df_clean
```

```python
def extract_answer(row):
    if pd.notnull(row['long_answers']):
        return row['long_answers'], "long"
    elif pd.notnull(row['short_answers']):
        return row['short_answers'], "short"
    else:
        return None, None
    

df_clean['answer'], df_clean['answer_type'] = zip(*df_clean.apply(extract_answer, axis=1))
df_clean[[ 'answer', 'answer_type']].head()
```

```python
# Check actual answer lengths
df_clean = df[df['long_answers'].notnull() | df['short_answers'].notnull()].copy()

# Get length statistics
df_clean['long_length'] = df_clean['long_answers'].fillna('').str.len()
df_clean['short_length'] = df_clean['short_answers'].fillna('').str.len()

print("SHORT ANSWERS:")
print(df_clean[df_clean['short_length'] > 0]['short_length'].describe())
print("\nLONG ANSWERS:")
print(df_clean[df_clean['long_length'] > 0]['long_length'].describe())

# Show some examples
print("\n--- SHORT ANSWER SAMPLES ---")
print(df_clean[df_clean['short_answers'].notnull()]['short_answers'].head(3).values)

print("\n--- LONG ANSWER SAMPLES ---")
print(df_clean[df_clean['long_answers'].notnull()]['long_answers'].head(3).values)
```

```python
# Check how many "short" answers are actually long
short_df = df_clean[df_clean['short_answers'].notnull()]
print("SHORT answers distribution:")
print(f"Under 100 chars: {(short_df['short_length'] < 100).sum()}")
print(f"100-500 chars: {((short_df['short_length'] >= 100) & (short_df['short_length'] < 500)).sum()}")
print(f"500-1000 chars: {((short_df['short_length'] >= 500) & (short_df['short_length'] < 1000)).sum()}")
print(f"Over 1000 chars: {(short_df['short_length'] >= 1000).sum()}")

# Check how many "long" answers are actually short
long_df = df_clean[df_clean['long_answers'].notnull()]
print("\nLONG answers distribution:")
print(f"Under 100 chars: {(long_df['long_length'] < 100).sum()}")
print(f"100-500 chars: {((long_df['long_length'] >= 100) & (long_df['long_length'] < 500)).sum()}")
print(f"500-1000 chars: {((long_df['long_length'] >= 500) & (long_df['long_length'] < 1000)).sum()}")
print(f"Over 1000 chars: {(long_df['long_length'] >= 1000).sum()}")
```
